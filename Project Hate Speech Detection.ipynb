```python
import re # re stands for regular expression. Regular expressions are a powerful tool for searching and manipulating strings. They can be used to find patterns in text, extract data, and perform other complex operations.
import nltk # nltk is a leading platform for building Python programs to work with human language data. stands for natural language tool kit
print(nltk.__version__)
import string # string is a module in Python that provides support for common string operations.
```

```python
nltk.download("stopwords") # Download stopwords.
```

```python
from nltk.corpus import stopwords # importing stopwords which is a part of nltk. used for removing stopwords.
stop_words = set(stopwords.words('english')) # The statement stop_words = set(stopwords.words('english')) creates a set of English stop words. Stop words are commonly used words that don't carry much meaning on their own, such as "the", "a", "an", "in", and "of".
```

```python
stemmer = nltk.stem.SnowballStemmer('english') # SnowballStemmer is an algorithm that reduces words to their root form, known as the stem.
```

```python
def clean_data(text):  # function to clean the data.
    text = str(text).lower() # Converts the text to lowercase.
    text = re.sub('https?://\S+|www\.S+', '', text) # Removes URLs.
    text = re.sub('\[.*?\]', '', text) # Removes text in square brackets.
    text = re.sub('<.*?>+', '', text) # Removes HTML tags.
    text = text + re.sub('[%s]'%re.escape(string.punctuation), '', text) # Removes punctuations. such as "!", "?", ".", ",", ";", ":", "-", "'", '"', and "(".
    text = re.sub('\n', '', text) # Removes new line characters.
    text = re.sub('\w*\d\w*', '', text) # Removes words containing numbers.
    text = re.sub(r"[^\w\s]", '', text)  # Remove special characters
    return text
```

```python
data1["tweet"] = data1["tweet"].apply(clean_data) # applies the clean_data function to the "tweet" column of the dataframe.
```

```python
data1
```

```python
x = np.array(data1["tweet"]) # "tweet" is the feature.
y = np.array(data1["labels"]) # "labels" is the target variable.
```

```python
x
```

```python
# The CountVectorizer class is used to convert a collection of text documents into a matrix of token counts.
# The sklearn.feature_extraction.text module in scikit-learn provides a number of tools for extracting features from text data. These features can then be used for various machine learning tasks such as text classification, clustering, and information retrieval.

from sklearn.feature_extraction.text import CountVectorizer #
from sklearn.model_selection import train_test_split

# The train_test_split function in scikit-learn is used to split a dataset into two subsets: a training set and a test set. This is a common practice in machine learning to evaluate the performance of a machine learning model.
# Syntax : train_test_split(X, y, test_size=0.25, train_size=None, random_state=None, shuffle=True, stratify=None)
```

```python
cv = CountVectorizer()
```

```python
x = cv.fit_transform(x) # transform() method of CountVectorizer class is used to convert a collection of text documents to a matrix of token counts.
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
x_train
```

```python
#Building ML model
from sklearn.tree import DecisionTreeClassifier # DecisionTreeClassifier is a machine learning algorithm that is used for classification tasks.
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train) # Fit the model to training data.
```

```python
y_pred = dt.predict(x_test) # Predict the target values for the test data.
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix # Confusion matrix function is used to check the accuracy of the model.
cr = confusion_matrix(y_test, y_pred)
cr
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
